### here doing some tranformation loading into silver layer

```python
from pyspark.sql.functions import *
```

```python
df=spark.table('practies.bronz.bronz_customer')\
     .withColumn('FirstName',upper(split(col('FullName'),' ')[0]))\
     .withColumn('LastName',upper(split(col('FullName'),' ')[1]))\
     .withColumn('Domine_Name',upper(split(col('Email'),'@')[0]))\
     .withColumn('Email',upper(col('Email')))\
      .drop("FullName")\
      .select("CustomerID","FirstName","LastName","Email","Domine_Name","PhoneNumber","JoinDate")

df.write.mode("overwrite")\
        .format("delta")\
        .saveAsTable("practies.silver_layer.silver_customer")
   
```

```python
df=spark.table('practies.bronz.bronz_Orders')\
     .withColumn('Status',upper(col('Status')))\
      .select("CustomerID","OrderID","TotalAmount","Status","OrderDate")

df.write.mode("overwrite")\
        .format("delta")\
        .saveAsTable("practies.silver_layer.silver_Orders")
```
