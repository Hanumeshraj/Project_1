```python
from pyspark.sql.functions import *
```

### creating customer table and order table in bronz layer

```python
df=spark.readStream.option("ignoreChanges","true")\
.table('workspace.default.customers')
df.writeStream.format('delta')\
      .outputMode('append')\
      .option('checkpointLocation', '/Volumes/practies/bronz/container/customers')\
    .trigger(availableNow=True)\
      .toTable('practies.bronz.bronz_customer')
```

```python
df=spark.readStream.table('workspace.default.orders')
df.writeStream.format('delta')\
      .outputMode('append')\
      .option('checkpointLocation', '/Volumes/practies/bronz/container/orders')\
    .trigger(availableNow=True)\
    .toTable('practies.bronz.bronz_orders')
```
